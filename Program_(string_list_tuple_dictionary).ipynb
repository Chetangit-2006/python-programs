```python
# Task 15: Reverse a given string using recursion

def reverse(s):
    if len(s) == 0:
        return s
    return reverse(s[1:]) + s[0]

print(reverse("Python"))
```

List Programs

```python
# Task 1: Write a Python program to get the largest number from a list

numbers = [10, 45, 23, 89, 67]   # List of numbers

largest = numbers[0]             # Assume first number is largest

for num in numbers:              # Loop through list
    if num > largest:            # Compare each number
        largest = num

print("Largest number is:", largest)   # Print result
```

```python
# Task 2: Write a Python program to print the numbers of a specified list after removing even numbers from it

numbers = [10, 15, 20, 25, 30, 35]   # Given list

result = []                         # Empty list to store odd numbers

for num in numbers:                 # Loop through list
    if num % 2 != 0:                # Check if number is odd
        result.append(num)          # Add odd number to new list

print("List after removing even numbers:", result)   # Print result
```

```python
# Task 3: Write a program to reverse a list in Python

numbers = [10, 20, 30, 40, 50]   # Original list

numbers.reverse()                # Reverse the list

print("Reversed list:", numbers)
```

```python
# Task 4: Python program to square each element of a list

numbers = [2, 3, 4, 5]     # Given list
result = []

for num in numbers:
    result.append(num ** 2)

print("Squared list:", result)
```

```python
# Task 5: Python program to remove an empty element from a list

num = ["Hello", 34, 45, "", 40]   # Given list

result = []

for item in num:
    if item != "":
        result.append(item)

print("Result:", result)
```

```python
# Task 6: Write a program to display items from a list that are divisible by 5

numbers = [10, 12, 15, 22, 25, 33, 40]

for num in numbers:
    if num % 5 == 0:
        print(num)
```

```python
# Task 7: Write a program to remove repetitive items from a list

num = [2, 3, 4, 5, 2, 6, 3, 2]   # Given list
result = []

for item in num:
    if item not in result:
        result.append(item)

print("Result:", result)
```

```python
# Task 8: Write a program to append data of the second list to the first list

list1 = [23, 24, 25, 26]
list2 = [27, 28, 29, 30]

list1.extend(list2)   # Append list2 into list1

print("Result:", list1)
```

```python
# Task 9: Write a program to filter odd and even numbers from a list

numbers = [2, 23, 24, 51, 46, 67]   # Given list

even = []
odd = []

for num in numbers:
    if num % 2 == 0:
        even.append(num)
    else:
        odd.append(num)

print("Even:", even)
print("Odd:", odd)
```

```python
# Task 10: Write a program to arrange elements in sorted order

numbers = [45, 12, 78, 34, 23]

numbers.sort()     # Sort list in ascending order

print("Sorted list:", numbers)
```

```python
# Task 11: Write a program to count the number of items in a list

numbers = [10, 20, 30, 40, 50]

count = len(numbers)

print("Total items:", count)
```

```python
# Task 12: Reverse the order of elements in the list using indexing method

chars = ['A', 'B', 'C', 'D', 'E']   # List of characters

reversed_list = []

for i in range(len(chars)-1, -1, -1):
    reversed_list.append(chars[i])

print("Reversed list:", reversed_list)
```

```python
# Task 15: Find the range (difference between largest and smallest element) of a list

numbers = [10, 45, 23, 89, 67]

largest = max(numbers)
smallest = min(numbers)

range_value = largest - smallest

print("Range of list:", range_value)
```

```python
# Task 16: Insert a new element at a specific position in the list

numbers = [10, 20, 30, 40, 50]   # Original list

numbers.insert(2, 25)           # Insert 25 at index 2

print("Updated list:", numbers)
```

```python
# Task 17: Find the common elements between two lists

list1 = [10, 20, 30, 40]
list2 = [30, 40, 50, 60]

common = []

for item in list1:
    if item in list2:
        common.append(item)

print("Common elements:", common)
```

```python
# Task 18: Find the second smallest element in a list

numbers = [45, 12, 78, 34, 23]

numbers.sort()                 # Sort the list

second_smallest = numbers[1]   # Second element after sorting

print("Second smallest element:", second_smallest)
```

```python
# Task 19: Remove elements at even indices from a list

numbers = [10, 20, 30, 40, 50, 60]

result = []

for i in range(len(numbers)):
    if i % 2 != 0:          # Keep elements at odd indices
        result.append(numbers[i])

print("Updated list:", result)
```

```python
# Task 20: Find the sum of elements at odd indices in a list

numbers = [10, 20, 30, 40, 50, 60]

total = 0

for i in range(len(numbers)):
    if i % 2 != 0:          # Odd index
        total += numbers[i]

print("Sum of elements at odd indices:", total)
```

Dictionary Program

```python
# 1. Add a key to a dictionary.

d = {0: 10, 1: 20}
d[2] = 30

print(d)
```

```python
# 2. Concatenate multiple dictionaries into a new one.

dic1 = {1:10, 2:20}
dic2 = {3:30, 4:40}
dic3 = {5:50, 6:60}

new_dic = {}
new_dic.update(dic1)
new_dic.update(dic2)
new_dic.update(dic3)

print(new_dic)
```

```python
# 3. Check whether a given key already exists in a dictionary.

d = {1: 10, 2: 20, 3: 30}

key = 2

if key in d:
    print("Key exists")
else:
    print("Key does not exist")
```

```python
# 4. Iterate over a dictionary using for loop.

d = {1: 10, 2: 20, 3: 30}

for key, value in d.items():
    print(key, value)
```

```python
# 5. Generate dictionary (x, x*x) between 1 and n.

n = 5
d = {}

for x in range(1, n+1):
    d[x] = x * x

print(d)
```

```python
# 6. Print dictionary with keys 1 to 15 and square values.

d = {}

for x in range(1, 16):
    d[x] = x * x

print(d)
```

```python
# 7. Merge two Python dictionaries.

d1 = {'a': 1, 'b': 2}
d2 = {'c': 3, 'd': 4}

d1.update(d2)
print(d1)
```

```python
# 8. Iterate over dictionary using for loop.

d = {'a': 10, 'b': 20, 'c': 30}

for key, value in d.items():
    print(key, value)
```

```python
# 9. Sum all items in a dictionary.

d = {'a': 10, 'b': 20, 'c': 30}

total = sum(d.values())
print(total)
```

```python
# 10. Multiply all items in a dictionary.

d = {'a': 2, 'b': 3, 'c': 4}

result = 1
for value in d.values():
    result *= value

print(result)
```

```python
# 11. Remove a key from a dictionary.

d = {'a': 1, 'b': 2, 'c': 3}

d.pop('b')
print(d)
```

```python
# 12. Map two lists into a dictionary.

keys = ['a', 'b', 'c']
values = [1, 2, 3]

d = dict(zip(keys, values))
print(d)
```

```python
# 13. Sort dictionary by key.

d = {3: 30, 1: 10, 2: 20}

for key in sorted(d):
    print(key, d[key])
```

```python
# 14. Get maximum and minimum values of a dictionary.

d = {'a': 5, 'b': 12, 'c': 3}

print("Maximum value:", max(d.values()))
print("Minimum value:", min(d.values()))
```

```python
# 15. Get dictionary from an object's fields.

class Student:
    def __init__(self):
        self.name = "Aryan"
        self.age = 20

s = Student()
print(s.__dict__)
```

```python
# 16. Remove duplicate values from dictionary.

d = {'a': 1, 'b': 2, 'c': 1, 'd': 3}

new_d = {}
for key, value in d.items():
    if value not in new_d.values():
        new_d[key] = value

print(new_d)
```

```python
# 17. Check if dictionary is empty or not.

d = {}

if not d:
    print("Dictionary is empty")
else:
    print("Dictionary is not empty")
```

```python
# 18. Combine two dictionaries by adding values for common keys.

from collections import Counter

d1 = {'a': 100, 'b': 200, 'c': 300}
d2 = {'a': 300, 'b': 200, 'd': 400}

result = Counter(d1) + Counter(d2)
print(result)
```

```python
# 19. Print all distinct values in a dictionary list.

data = [{"V":"S001"}, {"V":"S002"}, {"VI":"S001"}, {"VI":"S005"},
        {"VII":"S005"}, {"V":"S009"}, {"VIII":"S007"}]

unique_values = set()

for item in data:
    for value in item.values():
        unique_values.add(value)

print("Unique Values:", unique_values)
```

```python
# 20. Create and display all combinations of letters,
# selecting each letter from a different key in a dictionary.

data = {'1': ['a', 'b'], '2': ['c', 'd']}

for i in data['1']:
    for j in data['2']:
        print(i + j)
```
